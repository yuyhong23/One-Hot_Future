```python
import pandas as pd
from functools import reduce
```

```python
emissions_goal = pd.read_csv('../clean_data/emission_goals.csv')
emissions_goal.head()
```

```python
china_allenergy = pd.read_csv('data_files/china_allenergy', index_col=0)
china_allenergy
```

```python
global_allenergy = pd.read_csv('data_files/global_allenergy', index_col=0)
india_allenergy = pd.read_csv('data_files/india_allenergy', index_col=0)
us_allenergy = pd.read_csv('data_files/us_allenergy', index_col=0)
eu_allenergy = pd.read_csv('data_files/eu_allenergy', index_col=0)
```

```python
data_frames = [china_allenergy, global_allenergy, india_allenergy, us_allenergy, eu_allenergy]
allenergy_combined = reduce(lambda  left,right: pd.merge(left,right,on=['years'],
                                            how='outer'), data_frames)
allenergy_combined
```

```python
allenergy_combined_T = allenergy_combined.T
allenergy_combined_T
```

```python
allenergy_combined_T= allenergy_combined_T.iloc[1:]
allenergy_combined_T
```

```python
allenergy_combined_T.reset_index(level=0, inplace=True)
```

```python
allenergy_combined_T.columns = ['country','2020_allenergy_predictions', '2030_allenergy_predictions', '2040_allenergy_predictions', '2050_allenergy_predictions']
allenergy_combined_T
```

```python
allenergy__predictions= pd.merge(emissions_goal, allenergy_combined_T, on='country' )
allenergy__predictions
```

```python
allenergy__predictions.to_csv('allenergy__predictions', encoding='utf-8')
```
