# Extract Data Into DataFrames For Nonrenewable Energy Types

### Data 1: bp.com Dataset on Oil Consumption (Exajoules)

```python
# Import data file
oil_file = "raw_data/bp-stats-review-2020-all-data.xlsx"

oil = pd.read_excel(oil_file, "Oil Consumption - EJ", engine="openpyxl")

# Rename the columns with value from the second row
oil_df = oil.rename(columns=oil.iloc[1])
display(oil_df.head(5))
display(oil_df.tail(5))
```

### Data 2: bp.com Dataset on Natural Gas Consumption (Exajoules)

```python
# Import data file
gas_file = "raw_data/bp-stats-review-2020-all-data.xlsx"

gas = pd.read_excel(gas_file, "Gas Consumption - EJ", engine="openpyxl")

# Rename the columns with value from the second row
gas_df = gas.rename(columns=gas.iloc[1])
display(gas_df.head(5))
display(gas_df.tail(5))
```

### Data 3: bp.com Dataset on Coal Consumption (Exajoules)

```python
# Import data file
coal_file = "raw_data/bp-stats-review-2020-all-data.xlsx"

coal = pd.read_excel(coal_file, "Coal Consumption - EJ", engine="openpyxl")

# Rename the columns with value from the second row
coal_df = coal.rename(columns=coal.iloc[1])
display(coal_df.head(5))
display(coal_df.tail(5))
```

### Data 4: bp.com Dataset on Nuclear Energy Consumption (Exajoules)

```python
# Import data file
nuclear_energy_file = "raw_data/bp-stats-review-2020-all-data.xlsx"

nuclear_energy = pd.read_excel(nuclear_energy_file, "Nuclear Consumption - EJ", engine="openpyxl")

# Rename the columns with value from the second row
nuclear_energy_df = nuclear_energy.rename(columns=nuclear_energy.iloc[1])
display(nuclear_energy_df.head(5))
display(nuclear_energy_df.tail(5))
```

### Data 5: bp.com Dataset on Hydroeletric Consumption (Exajoules)

```python
# Import data file
hydroelec_file = "raw_data/bp-stats-review-2020-all-data.xlsx"

hydroelec = pd.read_excel(hydroelec_file, "Hydro Consumption - EJ", engine="openpyxl")

# Rename the columns with value from the second row
hydroelec_df = hydroelec.rename(columns=hydroelec.iloc[1])
display(hydroelec_df.head(5))
display(hydroelec_df.tail(5))
```
