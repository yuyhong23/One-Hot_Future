#### Nonrewable energy consumption (Exajoules)

```python
# Subtract primary energy consumption df by renewable energy consumption
bp_nonrewablec = bp_ec_all_cleaned.subtract(bp_rc_all_cleaned)
bp_nonrewablec
```

```python
# Divide nonrenewable energy consumption by primary energy consumption df
bp_nonrewablec_div = bp_nonrewablec.divide(bp_ec_all_cleaned)
bp_nonrewablec_div
```

```python
# Multiple by 100
bp_nonrewablec_div = bp_nonrewablec_div *100
bp_nonrewablec_div
```

```python
# Divide renewable energy consumption by primary energy consumption df
bp_rewablec_div = bp_rc_all_cleaned.divide(bp_ec_all_cleaned)
bp_rewablec_div
```

```python
# Multiple by 100
bp_rewablec_div = bp_rewablec_div *100
bp_rewablec_div
```

```python
# Add suffix to all years for all energy consumption
bp_ec_all_cleaned = bp_ec_all_cleaned.add_suffix('_all_energy')
bp_ec_all_cleaned
```

```python
# Add suffix to all years for renewable energy consumption
bp_rewablec_div = bp_rewablec_div.add_suffix('_renewable%')
bp_rewablec_div
```

```python
# Add suffix to all years for renewable energy consumption
bp_nonrewablec_div = bp_nonrewablec_div.add_suffix('_nonrenewable%')
bp_nonrewablec_div
```

#### Dataframes Combination

```python
# Combine all energy dataframe with renewable dataframe
data_combine1 = pd.concat([bp_ec_all_cleaned, bp_rewablec_div], axis=1)
data_combine1
```

```python
# Combine with nonrenewable dataframe
data_combine = pd.concat([data_combine1, bp_nonrewablec_div], axis=1)
data_combine
```
