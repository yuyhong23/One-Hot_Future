```python
import pandas as pd
from functools import reduce
```

```python
nonrenew = pd.read_csv('clean_data/nonrenewable_energy.csv')
nonrenew
```

```python
nonrenew.at[102, "country"] = "EU"
```

```python
nonrenew.at[99, "country"] = "Global"
nonrenewable_df = nonrenew[nonrenew["country"].isin(["Global", "US", "EU", "India", "China"])]
```

```python
# Flip the df
nonrenewT = nonrenewable_df.T
nonrenewT = nonrenewT.reset_index(drop=False)
# Rename columns
nonrenewT = nonrenewT.rename(columns={"index":"year", 99: "oil"})
# all_energyT.set_index("year", inplace=True)
nonrenewT

renew_final = nonrenewT.rename(columns={'oil' :'Global', 2:'US', 102:'EU', 84:'India', 82:'China'})

renew_final = renew_final.iloc[1:]
renew_final
```

```python
nonrenewable_global = pd.read_csv('goals_prediction_ML_tara/predictions_data_files/nonrenewable_global.csv', index_col=0)
nonrenewable_china = pd.read_csv('clean_data/china_unrenewableadjust_recent.csv', index_col=0)
nonrenewable_india = pd.read_csv('clean_data/india_unrenewable_adjust_recent.csv', index_col=0)
nonrenewable_us = pd.read_csv('clean_data/us_unrenewable_adjust_recent.csv', index_col=0)
nonrenewable_eu = pd.read_csv('clean_data/eu_unrenewable_adjust_recent.csv', index_col=0)
```

```python
nonrenewable_global = nonrenewable_global.rename(columns={"years":"year"})
```

```python
data_frames = [nonrenewable_eu, nonrenewable_global, nonrenewable_china, nonrenewable_india, nonrenewable_us]
nonrenew_combined = reduce(lambda  left,right: pd.merge(left,right,on=['year'],
                                            how='outer'), data_frames)
nonrenew_combined
```

```python
nonrenew_final_combine = renew_final.append(nonrenew_combined)
nonrenew_final_combine.set_index("year", inplace=True)
nonrenew_final_combine
```

```python
nonrenew_final_combine = nonrenew_final_combine[["Global", "US", "EU", "India", "China"]]
nonrenew_final_combine
```

```python
nonrenew_final_combine.to_csv('graph_data/nonrenewable_predictions_adjust.csv', encoding='utf-8')
```
